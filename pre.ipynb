```python
def convert(x):
    ids = np.unique(x)
    vals = 0
    pd = np.zeros((x.shape))
    for item in ids:
        pd[x==item] = vals
    return pd
```

```python
train = pd.read_csv("train.csv")
train.head()
```

```python
train1 = pd.read_csv("train1.csv")
train1.head()
```

```python
train2 = pd.read_csv("train2.csv")
train2.head()
```

```python
train3 = pd.read_csv("train3.csv")
train3.head()
```

```python
X1 = train1.drop(['Y'], axis=1)
X1['ID'] = convert(X1['ID'].values)
X1 = X1.apply(pd.to_numeric,errors='coerce')
X1.fillna(0,inplace =True)
X1=X1.astype(float)
pd.options.display.max_columns = None
X1
```

```python
X2 = train2.drop(['Y'], axis=1)
X2['ID'] = convert(X2['ID'].values)
X2 = X2.apply(pd.to_numeric,errors='coerce')
X2.fillna(0,inplace =True)
X2=X2.astype(float)
pd.options.display.max_columns = None
X2
```

```python
X3 = train3.drop(['Y'], axis=1)
X3['ID'] = convert(X3['ID'].values)
X3 = X3.apply(pd.to_numeric,errors='coerce')
X3.fillna(0,inplace =True)
X3=X3.astype(float)
pd.options.display.max_columns = None
X3
```

```python
y1=train1['Y']
y2=train2['Y']
y3=train3['Y']
y1,y2,y3
```

```python
X = train.drop(['Y'], axis=1)
X['ID'] = convert(X['ID'].values)
X = X.apply(pd.to_numeric,errors='coerce')
X.fillna(0,inplace =True)
X=X.astype(float)
pd.options.display.max_columns = None
X
```

```python
y=train['Y']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
clf = GradientBoostingClassifier(loss="deviance",learning_rate=0.1,n_estimators=225,max_features=46,min_samples_split=75,min_samples_leaf=1,random_state=0)
clf.fit(X_train, y_train)
test_predict=clf.predict(X_test)
```
